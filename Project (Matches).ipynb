```python
types=matches['city'].value_counts()
types.reset_index()
```

```python
plt.pie(types,labels=types.index,colors=['r','g','b','c','y'])
plt.show()
```

```python
cat_cols=matches.select_dtypes(include=['object']).columns
num_cols=matches.select_dtypes(include=np.number).columns.tolist()
print("Categorial Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew:',round(matches[col].skew(),2))
    plt.figure(figsize=(15,4))
    plt.subplot(1,2,1)
    matches[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1,2,2)
    sns.boxplot(x=matches[col])
    plt.show()
```
